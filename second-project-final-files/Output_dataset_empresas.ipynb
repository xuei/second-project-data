```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel ("Dataset_empresas.xls").reset_index()
df.head()
```

```python
# Header change

df.columns = ["Index","Region","Industry Type", "Nº Companies", "Total Salaries €", "Net Result", "Nº Employees", "Total Turnover €", "Operational Profitability %", "Survival rate %"]

df.head()
```

```python
# How many lines per industry type will be affected by the data cleaning

df1 = df.groupby("Industry Type")["Nº Companies"].sum()

df.isnull().sum()
```

```python
# Applying the Data Cleaning Axis 0 (by default) to affect rows

df2 = df.dropna()
df2.head()
```

```python
# analise qualidade de vida 

life = pd.read_excel ("Life.xls")

life['Crimes/Population']= -life['Registered Crimes']/ life['Population']

label= np.arange(3)+1

life['Purchasing power']= [int(value) for value in pd.qcut(life['Purchasing power'],3, labels= label).values ]

life['Environment Expenses']= [int(value) for value in pd.qcut(life['Environment Expenses'],3, labels= label).values ]
life['Gross Income']= [int(value) for value in pd.qcut(life['Gross Income'],3, labels= label).values ]
life['Crimes/Population']= [int(value) for value in pd.qcut(life['Crimes/Population'],3, labels= label).values ]

life['Quality of Life']= life['Purchasing power'] + life['Environment Expenses'] + life['Gross Income'] + life['Crimes/Population']


def quality(i):
    
    if i >=9:
        ft= 'Good Quality of Life'
        return ft
    elif i >=7:
        sc= 'Moderate Quality of Life' 
        return sc  
    else: 
        td ='Low Quality of Life'
        return td

life['Quality of Life']= life['Quality of Life'].apply(quality)
df_qualidade = life.rename(columns={'Unnamed: 0': 'Region'})
df_qualidade

```
